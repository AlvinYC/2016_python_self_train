```python
#encoding utf-8
import numpy as np
import pandas as pd
from IPython.display import Image
```

```python
pd.set_option("display.max_rows",140)
DTFQA = pd.read_pickle('./db/DTFQA_v03')
#DTFQA.iloc[0:100,:]
DTFQA
```

```python
pos_list  = []
term_list = []
for idx in range(len(DTFQA['pos'])):
    pos_list  += DTFQA['pos'][idx]
    term_list += DTFQA['term'][idx]
```

```python
type(DTFQA['pos'][0])
```

```python
print 'pos len = ' + str(len(pos_list))  
print pos_list[0:10] 
print 'term len = ' + str(len(term_list)) 
print repr(term_list[0:10]).decode("unicode-escape")
```

```python
# 將 pos/term row data 製作成 dataframe
data = {'pos': pos_list, 'term': term_list}
DTFQA_STATISTIC = pd.DataFrame(data)
DTFQA_STATISTIC.iloc[0:10,:]
```

```python
len(DTFQA_STATISTIC)
```

```python
# 計算取得所有 pos 的個數 (term 內容包含重複)
#DTFQA_STATISTIC.groupby(['pos','term']).count()
POS_COUNT = DTFQA_STATISTIC.groupby('pos').count()
print 'number of all pos type is ' + str(len(POS_COUNT))
#POS_COUNT
POS_COUNT_SORT = POS_COUNT.sort(['term'], ascending=[False])
POS_COUNT_SORT
```

```python
DTFQA_STATISTIC.groupby(['pos']).count()
```

```python
DTFQA_STATISTIC.groupby(['pos','term']).count()
```

```python
######### output
DTFQA_STATISTIC.to_pickle('./db/DTFQA_STATISTIC')
```

```python
#取得所有 term 的所有可能 (移除重複)

pd.set_option("display.max_rows",300)
POS_DETAIL = pd.DataFrame({'count' : DTFQA_STATISTIC.groupby( [ "pos", "term"] ).size()}).reset_index()
print 'all kind of term == ' + str(len(POS_DETAIL))
POS_DETAIL
```

```python
#對某一種 pos 查詢其所有可能 (移除重複)
target_idx = np.where(POS_DETAIL['pos'] == 'notdefine')
POS_TARGET = POS_DETAIL.iloc[target_idx[0],:]
POS_SORT   = POS_TARGET.sort(['count'], ascending=[False])
print 'len of POS_SORT for \'n\' = ' + str(len(POS_SORT['pos']))
POS_SORT
```
