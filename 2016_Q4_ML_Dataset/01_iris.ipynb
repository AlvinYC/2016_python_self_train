```python
X = df.iloc[0:100,[0,2]].values
X
```

```python
plt.scatter(X[:50,0],X[:50,1],color='r',marker='o',label='setosa')
plt.scatter(X[50:100,0], X[50:100,1],color='b',marker='x',label='versicolor')
plt.xlabel('sepal len(cm)')
plt.ylabel('petal len(cm)')
plt.legend(loc='upper left')
plt.show()
# 結論
# 簡單地說, setosa 不管是花萼還是花瓣 長度都比 vericolor 短
```

```python
X = df.iloc[0:100,[1,3]].values
plt.scatter(X[:50,0],X[:50,1],color='r',marker='o',label='setosa')
plt.scatter(X[50:100,0], X[50:100,1],color='b',marker='x',label='versicolor')
plt.xlabel('sepal width(cm)')
plt.ylabel('petal width(cm)')
plt.legend(loc='upper left')
plt.show()
# 結論
# 但是以寬度來說, setosa 的花萼寬度越寬時 花瓣比較短
# 而 versicolor 顛倒, 在花萼越窄時,花辦的寬度越寬
```

```python
#如果我們看三個類時,還能這麼容易分辨嗎
X = df.iloc[0:150,[0,2]].values
plt.scatter(X[:50,0],X[:50,1],color='r',marker='o',label='setosa')
plt.scatter(X[50:100,0], X[50:100,1],color='b',marker='x',label='versicolor')
plt.scatter(X[100:150,0], X[100:150,1],color='g',marker='s',label='virginica')

plt.xlabel('sepal len(cm)')
plt.ylabel('petal len(cm)')
plt.legend(loc='upper left')
plt.show()
```
