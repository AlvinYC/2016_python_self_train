```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 2
menMeans = (48,52)

ind = np.arange(N)    # the x locations for the groups
width = 0.6       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, menMeans, width,color=('r','y'),align='center' )

plt.ylabel('number of uttrance')
plt.title('Testing utterance distribution')
plt.xticks(range(len(menMeans)), ('Find Restaurant','the others'))
plt.yticks(np.arange(0, 81, 10))
#-------
plt.text(0-0.05,menMeans[0]/2, '48',fontsize=20)
plt.text(1-0.05,menMeans[1]/2, '52',fontsize=20)
plt.legend((p1[0], p1[1]), ('Find Restaurant', 'the others'))

plt.show()
```

```python
N = 2
menMeans = (80,69)

ind = np.arange(N)    # the x locations for the groups
width = 0.6       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, menMeans, width,color=('r','y'),align='center' )

plt.ylabel('predict as \'find resturant\'')
plt.title('Intent prediction as \'find restaurant\'')
plt.xticks(range(len(menMeans)), ('MS LUIS','CEI NLP'))
plt.yticks(np.arange(0, 101, 10))
#-------
plt.text(0-0.05,menMeans[0]/2, '81',fontsize=20)
plt.text(1-0.05,menMeans[1]/2, '69',fontsize=20)
plt.legend((p1[0], p1[1]), ('MS LUIS', 'CEI NLP'))

plt.show()
```

```python
N = 2
menMeans = (80,69)
FindRest = (47,45)

ind = np.arange(N)    # the x locations for the groups
width = 0.6       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, menMeans, width,color=('r','brown'),align='center' )
p2 = plt.bar(ind, FindRest, width,color=('yellowgreen','yellowgreen'),align='center' )


plt.ylabel('predict as \'find resturant\'')
plt.title('Intent prediction as \'find restaurant\'')
plt.xticks(range(len(menMeans)), ('MS LUIS','CEI NLP'))
plt.yticks(np.arange(0, 101, 10))
#-------
plt.text(0-0.05,menMeans[0]*1.02, '81',fontsize=20)
plt.text(1-0.05,menMeans[1]*1.02, '69',fontsize=20)
plt.text(0-0.28,menMeans[0]*4/5, '34(41.9%)',fontsize=20)
plt.text(1-0.28,menMeans[1]*4/5, '24(34.7%)',fontsize=20)
plt.text(0-0.05,menMeans[0]/5, '47',fontsize=20)
plt.text(1-0.05,menMeans[1]/5, '45',fontsize=20)
plt.legend((p1[0], p1[1]), ('MS LUIS', 'CEI NLP'))

plt.show()
```
